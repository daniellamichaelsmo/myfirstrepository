```python
#flag revenue over a certain size
flags = []

for rev in df["revenue"]:
    if rev >= 200:
        flags.append("large")
    else:
        flags.append("small")
        
```

```python
flags
df["order_size_flags"] =  flags
```

```python
df
```

```python
#functions pd examples
df["revenue"] >= 200
```

```python
#np.where
df["order_size_np"] = np.where(
    df["revenue"] >= 200,
    "large",
    "small"
)
```

```python
df
```

```python
df["revenue_adj"] = df["revenue"]
```

```python
#looping by index
for i, rev in enumerate(df["revenue"]): #index and value accessed by enumerate
    if i % 4 ==0:
        df["revenue_adj"][i] = df["revenue"][i] * 1.1
```

```python
rev_by_cat = df.groupby("category")["revenue"].sum()
```

```python
rev_by_cat_region = df.groupby(["category", "region"])["revenue"].mean()
```

```python
#reset index method
rev_by_cat_region = rev_by_cat_region.reset_index()
rev_by_cat_region
```

```python
#agg method .agg
cat_summary = df.groupby("category").agg(
    total_revenue=("revenue", "sum"),
    line_count=("order_id", "count"),
    avg_quantity=("quantity", "mean")
)
cat_summary = cat_summary.reset_index()
cat_summary
```

```python
#map method
df["return_status"] = df["returned"].map({True: "Returned", False: "Not Returned"})
```

```python
df[["returned", "return_status"]]
```

```python
#cut method
df["price_band"] = pd.cut(
    df["unit_price"],
    bins=[0,15,50,200],
    labels=["low", "mid", "High"]
)

df[["unit_price", "price_band"]]
```

```python
#visualize "What category brings in the most revenue"
#Make a Bar Chart
rev_by_cat = df.groupby("category")["revenue"].sum()

rev_by_cat.plot(kind="bar")
plt.title("Total Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
#Rev by return and not return
#grouped bar chart
return_plot = df.groupby(
    ["category", "return_status"])["revenue"].sum().unstack()

return_plot.plot(kind="bar")
plt.title("Revenue by Category and Return Status")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
#revenue over time by category
daily = (df.groupby(["order_date", "category"])["revenue"].sum().reset_index())

daily_wide =daily.pivot(index="order_date", columns="category", values="revenue").fillna(0)

daily_wide.plot(kind="line")
plt.title("Daily Revenue Over Time by Category")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```

```python
import matplotlib.pyplot as plt

# ensure datetime
df["order_date"] = pd.to_datetime(df["order_date"])

daily = (
    df.groupby(["order_date", "category"])["revenue"]
      .sum()
      .reset_index()
)

daily_wide = (
    daily.pivot(index="order_date", columns="category", values="revenue")
         .fillna(0)
         .sort_index()
)

ax = daily_wide.plot(figsize=(12, 6), linewidth=2, alpha=0.85)

plt.title("Daily Revenue Over Time by Category")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

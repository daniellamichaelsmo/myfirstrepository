```python
students = students.drop_duplicates()
```

```python
students["first_name"] = students["first_name"].str.strip().str.upper()
```

```python
students["last_name"] = students["last_name"].str.strip().str.upper()
```

```python
students["major"].str.strip().str.capitalize()
```

```python
students["major"] = students["major"].str.strip().str.capitalize()
students["major"] = students["major"].replace({
    "Comp sci" : "Computer Science",
    "Cs" : "Computer Science",
    "Computer science" : "Computer Science",
    "Bio" : "Biology",
    "Psych" : "Psychology"
    })
```

```python
students["major"].value_counts()
```

```python
students["course_grade"] = students["course_grade"].str.strip()
students.loc[students["course_grade"] == "ninety", "course_grade"] = 90
```

```python
students["course_grade"].value_counts()
```

```python
students["age"].describe()
```

```python
students.loc[students["age"] < 5, "age"] = np.nan
```

```python
students.loc[students["age"] > 100, "age"] = np.nan
```

```python
#first = last name challenge
students["full_name"] = students["first_name"] + ' ' + students["last_name"]
```

```python
students["full_name"]
```

```python
students["age_group"] = pd.cut(students["age"],
     bins=[0, 25, 40, 65,],
     labels=["Young adults", "working adults", "Retired"]
     )
```

```python
students[["age", "age_group"]].value_counts(dropna=False)
```

```python
students['cost_per_semester'] = students['tuition_paid'] / 3
```

```python
students['cost_per_semester'].describe()
```

```python
students["enrollment_date"] = pd.to_datetime(students["enrollment_date"])   
```

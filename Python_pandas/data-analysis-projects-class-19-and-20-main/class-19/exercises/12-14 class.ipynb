```python
#make df for fitness_data.csv
df = pd.read_csv("fitness_data.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column="resting_hr", by="activity_level")
plt.title()
plt.suptitle("")
plt.xlable()
plt.ylable()
plt.show()
```

```python

#days where stress is higher than 70
high_stress = df[df["stress_score"]> 60]
'''
# Plotting
plt.figure(figsize=(12,6))
plt.plot(high_stress["Day_of_week",] high_stress["stress_score"], marker='o', linestyle='-', color='red')
plt.title("High Stress Days (Stress Score > 70)")
plt.xlabel("Day_of_Week")
plt.ylabel("Stress Score")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
'''


```

```python
plt.scatter(high_stress["sleep_hours"], high_stress["resting_hr"], color="blue", alpha = 0.6)
plt.title("Sleep Hours vs Resting Heart Rate on High Stress Days")
plt.xlabel("Sleep Hours")
plt.ylabel("Resting Heart Rate")
plt.xlim(5,9.5)
plt.ylim(57,70)
plt.show()
```

```python
jan = df[df["date"] < "2024-02-01"]
fig, axes = plt.subplots(2,1, figsize=(10,8), sharex=True)

axes[0].plot(jan["date"], jan["resting_hr"], color="darkred")
axes[0].set_title("Resting Heart Rate Over Time")

axes[1].plot(jan["date"], jan["sleep_hours"], color="blue")
axes[1].set_title("Resting Heart Rate Over Time")

plt.xticks(rotation=45)
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,5), sharex=True, sharey=True)

levels=["Low", "Moderate", "High"]
colors=["green", "orange", "purple"]
for ax, level, color in zip(axes, levels, colors):
    subset = df[df["activity_level"]==level]
    ax.scatter(subset["steps"], subset["calories_burned"], color=color, alpha=0.6)
    ax.set_title(f"Activity Level: {level}")
    ax.set_xlabel("Steps")
    ax.set_ylabel("Calories Burned")
```

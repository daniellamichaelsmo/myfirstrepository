## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#Remove rows with missing city or menu amount (renaming this to price in step 3)
df_clean = Restaurant_data_df.dropna(subset=['menus.amountMax', 'city'])

#Aggregate average menu amount by city
avg_amount_by_city = df_clean.groupby('city')['menus.amountMax'].mean()

#Select top 25 cities
top25_cities = avg_amount_by_city.sort_values(ascending=False).head(25)

#Plot
plt.figure(figsize=(12,6))
top25_cities.plot(kind='bar', color='skyblue')
plt.title('Top 25 Cities by Average Menu Amount (Max)')
plt.xlabel('City')
plt.ylabel('Average Menu Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Number of Items by Province
#(Name changed From Province to State in checkpoint 3)
#Remove rows with missing province or menu name
df_clean = Restaurant_data_df.dropna(subset=['province', 'menus.name'])

#Count number of menu items per province
items_per_province = df_clean.groupby('province')['menus.name'].count()

#Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(items_per_province.index, items_per_province.values, color='purple')
plt.title('Number of Menu Items by Province')
plt.xlabel('Province')
plt.ylabel('Number of Menu Items')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
